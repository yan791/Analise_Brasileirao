```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Dados = pd.read_csv('brasileirao.csv', encoding='latin-1')
gols_time = Dados.groupby('team')['goals'].sum()
gols_time = gols_time.sort_values(ascending = False)
gols_time.head()
```

```python
plt.figure(figsize = (10, 4))
gols_time.head(10).plot(kind = 'bar') 
plt.title('Times com mais gols no brasileirao')
plt.xticks(rotation = 45)
plt.xlabel('')
plt.ylabel('Quantidade de gols')
plt.show()
```
