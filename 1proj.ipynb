```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
Dados = pd.read_csv('brasileirao.csv', encoding='latin-1')
gols_time = Dados.groupby('team')['goals'].sum()
gols_time = gols_time.sort_values(ascending = False)
gols_time.head()
```

```python
Dados.describe()
```

```python
plt.figure(figsize = (10, 4))
gols_time.head(10).plot(kind = 'bar') 
plt.title('Times com mais gols no brasileirao')
plt.xticks(rotation = 45)
plt.xlabel('')
plt.ylabel('Quantidade de gols')
plt.show()
```

```python
media = Dados.groupby('season')['goals'].mean()
plt.figure(figsize = (12, 5))
media.plot(kind = 'line')
plt.title('Media de gols por temporada')
plt.xlabel('Anos')
plt.ylabel('Quantidade de gols')
plt.show()
```
