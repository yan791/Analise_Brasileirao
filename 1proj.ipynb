```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
Dados = pd.read_csv('brasileirao.csv', encoding='latin-1')
```

```python
Dados.describe()
```

```python
gols_time = Dados.groupby('team')['goals'].sum().sort_values(ascending = False)
plt.figure(figsize = (10, 4))
gols_time.head(10).plot(kind = 'bar')
plt.title('Times com mais gols no brasileirao')
plt.xlabel('')
plt.xticks(rotation = 45)
plt.ylabel('Quantidade de Gols')
plt.show()
```

```python
media = Dados.groupby('season')['goals'].mean()
plt.figure(figsize= (10, 4))
media.plot(kind = 'line')
plt.title('Media de gols por temporada')
plt.xlabel('Ano')
plt.ylabel('Quantidade de Gols')
plt.show()
```
